```python
from statsmodels.formula.api import ols
import seaborn as sns

sns.get_dataset_names()
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
form = 'mpg ~ cylinders + displacement + horsepower + weight + acceleration'
model = ols(form, data=mpg).fit()
model.summary()
```
