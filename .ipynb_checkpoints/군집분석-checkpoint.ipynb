### 군집분석
- 비지도 학습
- 종류
    - 계층적 군집: 군집의 수를 미리 정하지 않고 군집을 구함. 거리를 기준으로 비슷한 군집끼리 묶음
        - 분할 방법
        - 응집 방법
    - 비계층적 군집:군지브이 수를 미리 정한 뒤 군집화
        - k-mean 클러스터링
        - 계층적 군집(Hierarchical Clustering)
       

### K-mean 클러스터링

```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
iris = sns.load_dataset('iris')
iris_x = iris.iloc[:,:-1]
iris_y = iris['species']
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, init='random')
model.fit(iris_x)
model.labels_
```

```python
pred = model.predict(iris_x)
```

```python
pd.crosstab(pred, iris_y)
```

```python
label = LabelEncoder()
label.fit(iris.species)

pred_str = label.inverse_transform(pred) # 학습시킨 모델을 통해 나온 결과값을 다시 역인코딩해줌

import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.subplot(121)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue=pred_str, style=pred_str)
plt.subplot(122)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', style='species')
```

## 계층적 클러스터링
- 덴드로그램을 활용하여 시각화

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
label.fit(iris.species)
iris.species = label.transform(iris.species)
```

```python
from scipy.cluster.hierarchy import linkage
cluster_model = linkage(iris, method='complete') # complete: 최장연결법, single: 최단연결법, average: 평균연결법
```

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(12,6))
dendrogram(cluster_model, labels= iris.index)
plt.show()
```

### 군집의 수 결정하기

```python
from scipy.cluster.hierarchy import fcluster
predict = fcluster(cluster_model, 4, criterion='distance')
```

```python
predict_adj = np.choose((predict-1),[0,2,1])
predict_origin = label.inverse_transform(predict_adj)
origin_name = label.inverse_transform(iris.species.values)
```

```python
pd.crosstab(predict_origin, origin_name, rownames=['Predict'],colnames=['TRUE'], margins=True)
```
