```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import minmax_scale, robust_scale, scale, maxabs_scale

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('csv/mtcars.csv')
```

```python
df.head()
```

## 정규화의 종류(메소드 이용)
- scale: 평균 0, 표준편차 1이 되도록 정규화
- robust_scale: 중위수, 사분위수 이용한 정규화
- minmax_scale: 최솟값 0 , 최댓값 1로 변수들 매핑
- maxabs_scale: 절대값이 가장 큰 값이 1, 작은값이 0, 0과 1사이의 값에 매핑, 음수는 부호 그대로 유지

```python
# 1. scale
scale_df = scale(df['qsec'])
scale_df
```

```python
sc_df = pd.DataFrame(scale_df, columns = ['scale'])
sc_df['var'] = range(sc_df.shape[0])
sc_df.head()
```

```python
plt.figure(figsize=(12,10))
plt.subplot(211)
sns.scatterplot(data=df, x= df['Unnamed: 0'], y=df['qsec'])

plt.subplot(212)
sns.scatterplot(data = sc_df, x=sc_df['var'], y= sc_df['scale'])
```

```python
## robust_scale()
rb_df = robust_scale(df['qsec'])
rb_df
```

```python
plt.figure(figsize=(12,10))
rb_df = pd.DataFrame(rb_df, columns = ['scale'])
rb_df['var'] = range(sc_df.shape[0])
sns.scatterplot(data = rb_df, x=sc_df['var'], y= sc_df['scale'])
```

```python
mms_df = minmax_scale(df['qsec'])

plt.figure(figsize=(12,10))
mms_df = pd.DataFrame(mms_df, columns = ['scale'])
mms_df['var'] = range(mms_df.shape[0])
sns.scatterplot(data = mms_df, x=mms_df['var'], y= mms_df['scale'])
```

```python
mas_df = maxabs_scale(df['qsec'])

plt.figure(figsize=(12,10))
mas_df  = pd.DataFrame(mas_df , columns = ['scale'])
mas_df ['var'] = range(mas_df .shape[0])
sns.scatterplot(data = mas_df , x=mms_df['var'], y= mms_df['scale'])
```

## 분류분석
- 분류 모델: mlpclassifier(), logistic
- 평가하기: accuracy, recall, precision, f1-score

```python
import seaborn as sns
iris=sns.load_dataset('iris')
iris_x = iris.iloc[:,:-1]
iris_y = iris['species']
train_x, test_x, train_y, test_y = train_test_split(iris_x, iris_y, test_size=.3)
```

```python
train_x.shape, test_x.shape
```

```python
from sklearn.metrics import r2_score, accuracy_score, recall_score, f1_score, precision_score

from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(50,30,20))
model = mlp.fit(train_x, train_y)

pred = model.predict(test_x)
pd.crosstab(test_y,pred)
```

### recall, precision, f1: y변수가 binary여야 함

```python
accuracy_score(test_y, pred)
```

### Classification_report

```python
from sklearn.metrics import classification_report

print(classification_report(test_y, pred))
```

### ROC curve를 이용한 모형 비교

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import roc_curve

model1 = LogisticRegression().fit(bc_x, bc_y)
model2 = SVC(gamma=.001, C=3000, probability = True).fit(bc_x, bc_y)

pred = model1.predict(bc_x)
pred2 = model2.predict(bc_x)

fpr1, tpr1, thr1 = roc_curve(bc_y, model1.decision_function(bc_x))
fpr2, tpr2, thr2 = roc_curve(bc_y, model2.decision_function(bc_x))

plt.plot(fpr1, tpr1, label='logistic')
plt.plot(fpr2, tpr2, label='SVM')
plt.legend()
plt.plot([0,1],[0,1],'k--')
```

### AUC

```python
from sklearn.metrics import auc
auc(fpr1,tpr1), auc(fpr2,tpr2)
```
